Question 1: What is a random variable in probability theory?
  - A random variable in probability theory is a variable whose value is a numerical outcome of a random phenomenon. It assigns numerical values to the possible outcomes of a random experiment.
  - For example, in a coin flip, the random variable could assign 1 for heads and 0 for tails. This allows us to quantify and analyze uncertain events.

Question 2: What are the types of random variables?
  - There are two main types of random variables:
    - Discrete random variable
     - Continuous rrandom variable

  1) Discrete random variables can only take on a finite number of values or a countably infinite number of values. These values are usually whole numbers.
  - Examples include the number of heads when flipping a coin multiple times or the number of cars passing a certain point on a road in an hour.

  2) Continuous random variables, on the other hand, can take on any value within a given range.
  - Examples include height, weight, or temperature.

Question 3: Explain the difference between discrete and continuous distributions.
  - Discrete distributions deal with random variables that can only take on distinct, separate values, often integers, like the number of heads in coin flips. They are described using probability mass functions (PMFs).
  
  - Continuous distributions involve variables that can take any value within a given range, such as height or temperature, and are described using probability density functions (PDFs). These functions calculate the probability of a variable falling within a specific interval.

Question 4: What is a binomial distribution, and how is it used in probability?

  - A binomial distribution models the probability of a specific number of successes in a fixed number of independent trials, where each trial has only two outcomes: success or failure.
  - It's used to calculate the probability of getting a certain number of successes, given the probability of success on a single trial.

Question 5: What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1.
  -  It's important because it serves as a reference point for all other normal distributions. By standardizing data (converting it to z-scores), we can compare values from different normal distributions and easily calculate probabilities using the standard normal table or calculator.

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that the distribution of sample means approximates a normal distribution, regardless of the original population's distribution, as the sample size increases.
  - It's critical in statistics because it allows us to make inferences about a population's mean, even if we don't know the population's distribution.

Question 7: What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals are significant in statistical analysis because they provide a range of values within which the true population parameter is likely to fall. They reflect the level of uncertainty around a sample estimate and help assess the precision and reliability of results. A 95% confidence interval, for example, means there's a 95% chance the interval contains the true value, helping in informed decision-making.

Question 8: What is the concept of expected value in a probability distribution?
  - The expected value in a probability distribution represents the average outcome we'd anticipate if we repeated an experiment many times. It's calculated by multiplying each possible outcome by its probability and summing these products. Essentially, it's the weighted average of all possible values, where the weights are the probabilities of each outcome. It gives us a sense of the "center" or "most likely" value you can expect.




Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
mean = 50
sd = 5
n = 1000

data = np.random.normal(loc=mean, scale=sd, size=n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=0)  # population std; used ddof=1 for sample std

print(f"Generated {n} random numbers from N({mean}, {sd**2})")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample std dev (population): {sample_std:.4f}")

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {sample_mean:.2f}')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.

```python
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
                235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_dev = np.std(daily_sales, ddof=1)
n = len(daily_sales)

standard_error = std_dev / np.sqrt(n)

# Z-score for 95% confidence level
z_score = stats.norm.ppf(0.975)

# Calculate margin of error
margin_of_error = z_score * standard_error

# Confidence interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
